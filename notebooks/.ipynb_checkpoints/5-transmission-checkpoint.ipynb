# Transmission amplitude at zero temperature

This notebook computes the transmission amplitude

$$ T_{LR}(t) = |\langle\Psi(t)|\Psi_R\rangle| \qquad \qquad T_{LL}(t) = |\langle\Psi(t)|\Psi_L\rangle|$$

where the states under consideration are Majorana excitations on the ground state $|G\rangle$ (so the system is at zero temperature)

$$ |\Psi_R\rangle = \chi_R|G\rangle 
        \qquad|\Psi_L\rangle = \chi_L|G\rangle 
        \qquad|\Psi(t)\rangle = e^{-iHt}\chi_R|G\rangle $$      
Time evolution is performed using the coupled Hamiltonian $H=H_L+H_R+H_{int}$. Physically, the square of the transmission amplitude $T_{LR}^2(t)$ reflects the the probability of recovering $\chi_L^j$ at some time $t$ after inserting $\chi_R^j$ at $t=0$, and similarly for $T_{LL}^2(t)$. For simplicity this notebook only computes $T_{LR}(t)$.

**Quantum many-body picture:** 

1. Start with ground state $|G\rangle$ of coupled SYK
2. Create excitation on Right system $|\Psi(t=0)\rangle = \chi_R|G\rangle$

3. Excitation gets scrambled
4. Excitation reassembles and becomes localized in Left system $|\Psi(t=t_\text{rev})\rangle = \chi_L|G\rangle$

5. Process is repeated with $L\leftrightarrow R$ $\Rightarrow$ Oscillatory behavior

**Dual gravitational picture:** Perturbation travels through the wormhole and reaches the Left boundary


```python
import numpy as np
from dynamite.extras import majorana
from dynamite.operators import op_sum, op_product
from dynamite.states import State
from dynamite.computations import eigsolve
from dynamite import config
from matplotlib import rcParams
import matplotlib.pyplot as plt

import sys
sys.path.append('..')
from Hypergraph import regular_hypergraph
from Hamiltonian import hamiltonian_coupled, get_gs
```

### Set parameters

```python
N = 8              # Number of Majoranas for single SYK 
k = 4              # Sparsity parameter
q = 4              # Sets q-body Hamiltonian
mu = 0.005         # Coupling stregth

tmax = 5000        
n_points = 500     
t_values = np.linspace(0, tmax, n_points)  # Values of time

config.L = N       # 2N Majorana -> N qubits
```

### Compute transmission amplitude

```python
def transmission_amplitude(N, k, q , hyperedges, mu, t_values, random_seed):
    '''
    Evaluate transmission amplitude as the overlap between \Psi_R(t) and \Psi_L
    '''
    
    # Build Hamiltonian and get ground state
    H = hamiltonian_coupled(N, k, q, hyperedges, mu, random_seed)
    G = get_gs(H)
    
    # Create single Majorana excitation on Right
    j = random_seed % N
    majR = majorana(N+j).scale(1 / np.sqrt(2))
    psiR = majR * G
    psiR.normalize()
    
    # Corresponding excitation on Left
    majL = majorana(j).scale(1 / np.sqrt(2))
    psiL = majL * G
    psiL.normalize()

    # Time evolution
    
    T_list = [abs(psiL.dot(psiR))]
                  
    for i in range(1, len(t_values)):
                  
        dt = t_values[i] - t_values[i-1]
        
        transmission = abs(psiL.dot(psiR))
        T_list.append(transmission) 
        
        H.evolve(psiL, t=-dt/2, result=psiL)
        H.evolve(psiR, t=dt/2, result=psiR)
        
    return np.array(T_list)
```

```python
%%time
hyperedges = regular_hypergraph(N, k, q, 0)
T = [transmission_amplitude(N, k, q, hyperedges, mu, t_values, s) for s in range(50)]
T_mean = np.mean(T, axis=0)
```

**Plot settings**

```python
rcParams['lines.linewidth'] = 2.0
rcParams['axes.linewidth'] = 2.0
plt.rc('font', size=18)          
plt.rc('axes', titlesize=18)     
plt.rc('axes', labelsize=22)   
plt.rc('xtick', labelsize=18) 
plt.rc('ytick', labelsize=18)   
plt.rc('legend', fontsize=18)  
plt.rc('figure', titlesize=24) 
plt.rcParams['mathtext.fontset'] = 'dejavuserif'
colors = ['cyan', 'green', 'red']
```

**Plot result**

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(1, 1, 1)

plt.plot(t_values, T[np.random.randint(50)], '-',  alpha=0.7, label='Single realization')
plt.plot(t_values, T_mean, '-', color='k', alpha=0.7, label='Average')

plt.title(r'$q={} \qquad \mu={} \qquad k={}$'.format(q, mu, k))
plt.xlabel(r'$t$')
plt.ylabel(r'$T_{LR}(t)$')
plt.xlim([0, tmax])
plt.ylim([0, 1])
plt.legend(loc='upper right')
plt.show()
```

### Projection into low-energy states

In order to enhance the transmission amplitude, we can consider the transmission of low-energy states, obtained as the projection of the Majorana excitation onto low-energy eigenstates of the coupled Hamiltonian
\begin{equation}
    |\Psi_R\rangle_\text{low-energy} = \frac{1}{\sqrt{C}}\sum_{n=0}^{n_\text{max}}e^{-a\frac{(E_n-E_0)}{|E_0|}}\langle \Psi_n|\chi^j_R|\Psi_0\rangle |\Psi_n\rangle,
\end{equation}
where $a$ is a suppression parameter, $n_max$ is the number of lowest eingestates we consider and $C$ is a normalization constant.

```python
# Evaluate projection of single Majorana excitation into low-energy components
def get_projection(H, majL, majR, a, nmax):
    
    # Get nmax low-energy eigenstates of Hamiltonian H
    E, psi = eigsolve(H, getvecs=True, nev=nmax, tol=1e-5)[:nmax]

    istart, iend = psi[0].vec.getOwnershipRange()
    
    projL = State()
    projR = State()
    projL.vec[istart:iend] = np.zeros(iend-istart).astype('complex')
    projR.vec[istart:iend] = np.zeros(iend-istart).astype('complex')

    for n in range(nmax):
        factor = psi[n].dot(majL * psi[0]) * np.exp(a * (E[n] - E[0]) / E[0])
        projL.vec[istart:iend] += psi[n].vec[istart:iend] * factor
        factor = psi[n].dot(majR * psi[0]) * np.exp(a * (E[n] - E[0]) / E[0])
        projR.vec[istart:iend] += psi[n].vec[istart:iend] * factor

    projL.vec.assemblyBegin()
    projL.vec.assemblyEnd()
    projR.vec.assemblyBegin()
    projR.vec.assemblyEnd()
    projL.normalize()
    projR.normalize()
    
    return projL, projR
```

### Evaluate low-energy transmission amplitude

```python
a_values = [0, 10, 100]  # Parameter of low-energy state
nmax = 10                # Number of eigenvectors to keep in low-energy state

# Build Hamiltonian
hyperedges = regular_hypergraph(N, k, q, 0)
H = hamiltonian_coupled(N, k, q, hyperedges, mu, 0)

# Select L and R Majorana fermions
majL = majorana(0).scale(1 / np.sqrt(2))
majR = majorana(N).scale(1 / np.sqrt(2))

T_low = [] # Store transmission amplitude

for a in a_values:
        
    # Create low-energy state
    psiL, psiR = get_projection(H, majL, majR, a, nmax)
    
    transmission = [abs(psiL.dot(psiR))]

    # evolve states in smalls steps dt/2 and compute overlap between them
    for i in range(1, len(t_values)):
        
        dt = t_values[i] - t_values[i-1]
            
        # Evaluate transmission amplitude
        transmission.append(abs(psiL.dot(psiR)))

        # Evolve states
        H.evolve(psiL, t=-dt/2, result=psiL)
        H.evolve(psiR, t=dt/2, result=psiR)
            
    T_low.append(transmission)
```

**Plot result**

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(1, 1, 1)

for i in range(len(a_values)):
    plt.plot(t_values, T_low[i], '-', 
             color=colors[i], 
             alpha=0.7, 
             label=r'$a={}$'.format(a_values[i]))

plt.title(r'$q={} \qquad \mu={} \qquad k={}$'.format(q, mu, k))
plt.xlabel(r'$t$')
plt.ylabel(r'$T_{LR}(t)$')
plt.xlim([0, t_values.max()])
plt.ylim([0, 1])
plt.legend(loc='upper right')
plt.show()
```
